# Face - face case

```python
def test_face(points, face_neighbors):   
    df = pd.DataFrame()
    df['xyz'] = points
    df['face_neighbors'] = face_neighbors
    df = create_face(df)
    df.insert(0, 'voxel_id', range(0, len(df)))
    df = calculate_face(df)
    cols = [i for i in range(3, 9)]
    df.drop(df.columns[cols],axis=1,inplace=True)
    
    return df
```

```python
# Across x, only 3 points
x0 = [-1, 0, 0]
x1 = [0, 0, 0]
x2 = [1, 0, 0]

points = [x0, x1, x2]
face_neighbors = [[1], [0, 2], [1]]

df_face_3points = test_face(points, face_neighbors)
```

```python
# Accross x, y, z only 27 points
x0 = [-1, 0, 0]
x1 = [-1, 0, 1]
x2 = [-1, 0, 2]
x3 = [-1, 1, 0]
x4 = [-1, 1, 1]
x5 = [-1, 1, 2]
x6 = [-1, 2, 0]
x7 = [-1, 2, 1]
x8 = [-1, 2, 2]

x9 = [0, 0, 0]
x10 = [0, 0, 1]
x11 = [0, 0, 2]
x12 = [0, 1, 0]
x13 = [0, 1, 1]
x14 = [0, 1, 2]
x15 = [0, 2, 0]
x16 = [0, 2, 1]
x17 = [0, 2, 2]

x18 = [1, 0, 0]
x19 = [1, 0, 1]
x20 = [1, 0, 2]
x21 = [1, 1, 0]
x22 = [1, 1, 1]
x23 = [1, 1, 2]
x24 = [1, 2, 0]
x25 = [1, 2, 1]
x26 = [1, 2, 2]

X = np.array([x0, x1, x2, x3, x4, x5, x6, x7, x8, x9, x10, x11, x12, x13,x14, x15,
             x16, x17, x18, x19, x20, x21, x22, x23, x24, x25, x26])

face_neighbors = []
for x in X:
    nei = []
    sub_loop = abs(x - X)
    for count, i in enumerate(sub_loop):
        if sum(i) == 1:
            nei.append(count)
    face_neighbors.append(nei)
    
df_face_27points = test_face(list(X), face_neighbors)
```
